```python
import folium 
import pandas as pd
import numpy as np
import pickle
import pycountry
import json
```

```python
geo_str = 'shp/custom.geo.json'

m = folium.Map(location=[48, -102], zoom_start=3)

d = pickle.load(open('countries_sentiment.p', 'rb'))
df =  pd.DataFrame.from_dict(d, orient = 'index')
df = df.reset_index()
df = df.rename(columns={"index": "name", 0: "sentiment"})
```

```python
folium.Choropleth(
    geo_data=geo_str,
    name='choropleth',
    data=df,
    columns=['name', 'sentiment'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Sentiment of News'
).add_to(m)

folium.LayerControl().add_to(m)

m
```
